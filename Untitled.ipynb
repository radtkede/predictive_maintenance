```python
import fbprophet
import pandas as pd
```

```python
df = pd.read_csv('/Users/dennis/git/Data/cleaned_2020.csv')
```

```python
df
```

```python
from fbprophet import Prophet
```

```python
m = Prophet(changepoint_prior_scale=0.01).fit(df)
future = m.make_future_dataframe(periods=300, freq='D')
fcst = m.predict(future)
fig = m.plot(fcst)
```

```python
from fbprophet.diagnostics import cross_validation
df_cv = cross_validation(m, initial='244 days', period='1 days', horizon = '3 days')
```
