```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.datasets import make_regression
from xgboost import XGBRegressor
from matplotlib import pyplot
%matplotlib inline 
from matplotlib import pyplot as plt
from matplotlib import style
import seaborn as sns
import string
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/Users/dennis/kaggle/input/predictive_maintenance/pump_data.csv', sep=';', low_memory=False , decimal=',')
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.drop(['Unnamed: 0'], axis = 1)
df = df.drop(['Date'], axis = 1)
df = df.drop(['03:YV4978'], axis = 1)
df = df.drop(['03:TC3550A'], axis = 1)
```

```python
df.head()
```

```python
df = df.replace(to_replace='N_LAUF', value='0')
df = df.replace(to_replace='LAUF', value='1')
df = df.replace(to_replace='No Data', value='-1001')
df = df.replace(to_replace='Over Range', value='-1002')
df = df.replace(to_replace='I/O Timeout', value='-1003')
df = df.replace(to_replace='Calc Failed', value='-1004')
df = df.replace(to_replace='Error', value='-1005')
df = df.replace(to_replace='Configure', value='-1006')
df = df.replace(to_replace='Scan Off', value='-1007')
df = df.replace(to_replace='Comm Fail', value='-1008')
df = df.replace(to_replace='Shutdown', value='-1009')
df = df.replace(to_replace='NaN', value='-1010')
df = df.replace(to_replace='Bad Input', value='-1011')
df = df.replace(to_replace='Bad', value='-1012')
```

```python
df.head()
```

```python
df.dtypes
dataset = df.replace(',','.', regex=True)
```

```python
dataset.head()
```

```python
for column in dataset:
    dataset[column] = pd.to_numeric(dataset[column], errors='coerce')

dataset.dtypes
```

```python
dataset['Durchsatz'] = dataset['03:FC3195'] + dataset['03:FC3197'] + dataset['03:FC3199']
```

```python
dataset = dataset.drop(['03:FC3195'], axis = 1)
dataset = dataset.drop(['03:FC3197'], axis = 1)
dataset = dataset.drop(['03:FC3199'], axis = 1)
```

```python
dataset.dropna(inplace=True)
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
print(dataset['Durchsatz'])
```

```python
corr = dataset.corr()
plt.subplots(figsize=(20,15))
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```

```python
X = dataset[dataset.columns[:-1]]
y = dataset[dataset.columns[-1]]
```

```python
model = XGBRegressor()
# fit the model
model.fit(X, y)
```
